```python
import pandas as pd
import  requests
from tqdm import tqdm
import time
```

```python
API_KEY = "8265bd1679663a7ea12ac168da84d2e8"

MOVIE_URL = "https://api.themoviedb.org/3/movie/top_rated"
GENRE_URL = "https://api.themoviedb.org/3/genre/movie/list"
```

```python
genre_response = requests.get(
    GENRE_URL,
    params={
        "api_key": API_KEY,
        "language": "en-US"
    }
)
```

```python
genre_data = genre_response.json()
```

```python
genre_dict = {}

for genre in genre_data["genres"]:
    genre_dict[genre["id"]] = genre["name"]

print(genre_dict)
```

```python
movies = []

for page in tqdm(range(1, 472)):   # 1 to 471

    response = requests.get(
        MOVIE_URL,
        params={
            "api_key": API_KEY,
            "language": "en-US",
            "page": page
        }
    )

    if response.status_code != 200:
        print(f"Error on page {page}")
        continue

    data = response.json()

    for movie in data["results"]:

        genre_names = []

        for gid in movie["genre_ids"]:
            if gid in genre_dict:
                genre_names.append(genre_dict[gid])

        movies.append({
            "movie_name": movie["title"],
            "description": movie["overview"],
            "genre": ", ".join(genre_names)
        })

    # Prevent hitting rate limits
    time.sleep(0.2)
```

```python
df=pd.DataFrame(movies)
df.head()
```

Removing html tags

```python
import re
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['description']=df['description'].apply(remove_html_tags)
```

```python
df = df.map(lambda x: remove_html_tags(x) if isinstance(x, str) else x)
```
